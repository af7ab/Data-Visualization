### Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

```python
# Calculate quartiles
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Define lower and upper bounds
lobound = Q1 - 1.5 * IQR
upbound = Q3 + 1.5 * IQR

# Remove outliers based on IQR bounds
updated_df = df[(df['Price'] >= lobound) & (df['Price'] <= upbound)]
updated_df

```

### You are a data analyst for a flight booking website, and you have been asked to analyze the Flight Price dataset to identify any trends in flight prices. What features would you analyze to identify these trends, and what visualizations would you use to present your findings to your team?

In order to identify price patterns, as a data analyst for travel websites, I would examine the journey date, airline, source, destination, and length of the flight. I would use line charts to show average prices over time, bar graphs to compare prices across airlines or destinations, heatmaps to look at price differences between cities, box plots to look at price distributions by month or season, and scatter plots to look at the relationship between price and length of flight. Decisions to optimize pricing strategies and enhance customer experience would be guided by these visualizations, which would show seasonal trends, competitive pricing tactics, and variations based on routes and flight lengths.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Flight Price dataset
df = pd.read_excel('flight_price.xlsx')
df['Date of Journey'] = pd.to_datetime(df['Date_of_Journey'])

# Visualize average prices over time for different airlines using line charts
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date_of_Journey', y='Price', hue='Airline', data=df)
plt.title('Average Prices Over Time for Different Airlines')
plt.xlabel('Date of Journey')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.legend(title='Airline', loc='upper left')
plt.show()

# Compare prices across airlines or destinations using bar graphs
plt.figure(figsize=(12, 6))
sns.barplot(x='Airline', y='Price', data=df)
plt.title('Average Prices Across Airlines')
plt.xlabel('Airline')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()

# Explore price variations between source and destination cities using heatmaps
plt.figure(figsize=(12, 8))
pivot_table = df.pivot_table(index='Source', columns='Destination', values='Price', aggfunc='mean')
sns.heatmap(pivot_table, cmap='viridis')
plt.title('Price Variations Between Source and Destination Cities')
plt.xlabel('Destination')
plt.ylabel('Source')
plt.show()


# Assess the relationship between flight duration and price using scatter plots
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Duration', y='Price', data=df)
plt.title('Relationship Between Flight Duration and Price')
plt.xlabel('Flight Duration')
plt.ylabel('Price')
plt.show()
```

### What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
import pandas as pd
import matplotlib.pyplot as plt


# Drop duplicate rows
df = df.drop_duplicates()

# Replace 'Varies with device' with NaN in 'Size' column
df['Size'] = df['Size'].replace('Varies with device', pd.NA)

# Convert 'Size' column to numeric (remove 'M' and 'k' suffixes)
df['Size'] = pd.to_numeric(df['Size'].replace('[kM]', '', regex=True), errors='coerce')

# Drop rows with missing values in 'Size' and 'Rating' columns
df_cleaned = df.dropna(subset=['Size', 'Rating'])

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['Size'], df_cleaned['Rating'],color = 'pink', alpha=0.5)
plt.title('Relationship between App Size and Rating')
plt.xlabel('App Size (MB)')
plt.ylabel('Rating')
plt.show()
```

### How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Drop duplicate rows
df = df.drop_duplicates()

# Replace 'Everyone' in 'Price' with NaN
df['Price'] = df['Price'].replace('Everyone', pd.NA)

# Convert 'Price' column to numeric (remove dollar sign)
df['Price'] = pd.to_numeric(df['Price'].replace('[\$,]', '', regex=True), errors='coerce')

# Filter only 'Free' and 'Paid' app types
df_filtered = df[df['Type'].isin(['Free', 'Paid'])]

# Group by 'Type' and calculate the average price for each app type
average_price_by_type = df_filtered.groupby('Type')['Price'].mean()

# Create a bar chart
plt.figure(figsize=(10, 6))
average_price_by_type.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()

```

```python
import pandas as pd

# Drop duplicate rows
df = df.drop_duplicates()

# Convert 'Installs' column to numeric
df['Installs'] = pd.to_numeric(df['Installs'].replace('[\+,]', '', regex=True), errors='coerce')

# Sort the DataFrame by 'Installs' in descending order
df_sorted = df.sort_values(by='Installs', ascending=False)

# Display the top 10 most popular apps
top_apps = df_sorted[['App', 'Installs']].head(10)
print("Top 10 Apps:")
print(top_apps)

# Create a frequency table for the names of the top 10 most popular apps
freq_table = top_apps['App'].value_counts().reset_index()
freq_table.columns = ['App', 'Frequency']

# Sort the frequency table by frequency in descending order
freq_table = freq_table.sort_values(by='Frequency', ascending=False)

# Display the frequency table
print("\nFrequency Table for Top 10 Apps:")
print(freq_table)
```

### A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

### To identify successful developers, I would examine the data from the Google Playstore and consider key metrics like size, category, downloads, user reviews, and app updates. Visualizations such as box plots for app size, scatter plots for user engagement, line charts for update frequency, bar graphs for downloads, and heatmaps for category distribution would effectively display these data. This study will help the mobile app development company grow by finding the top developers for potential collaborations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Google Playstore dataset
url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
playstore_data = pd.read_csv(url)

# Data preprocessing
playstore_data['Reviews'] = pd.to_numeric(playstore_data['Reviews'], errors='coerce')  # Convert 'Reviews' to numeric

# Analyze key features
# 1. Number of Downloads
total_downloads = playstore_data['Installs'].sum()
print("Total Number of Downloads:", total_downloads)

# 2. User Ratings and Reviews
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rating', y='Reviews', data=playstore_data)
plt.title('User Ratings vs Reviews')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.show()

# 3. App Updates
total_updates = playstore_data['Last Updated'].count()
print("Total Number of App Updates:", total_updates)

# 4. App Size
plt.figure(figsize=(10, 6))
sns.boxplot(x='Size', data=playstore_data)
plt.title('App Size Distribution')
plt.xlabel('App Size')
plt.ylabel('App Count')
plt.show()

# 5. Category and Genre
category_genre_pivot = playstore_data.pivot_table(index='Category', columns='Genres', values='App', aggfunc='count')
plt.figure(figsize=(12, 8))
sns.heatmap(category_genre_pivot, cmap='viridis')
plt.title('Category and Genre Distribution')
plt.xlabel('Genre')
plt.ylabel('Category')
plt.show()
```

#### Analyze the Google Playstore data thoroughly by looking at "Installs" for popularity, "Ratings" and "Reviews" for user satisfaction, and pricing distribution for different app categories. Make a graphic depiction of the results and recommend product categories that have a lot of traction, good customer feedback, and reasonable costs. Iterate and refine suggestions while accounting for other characteristics such as "Size," "Content Rating," and "Genres."

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data Cleaning
df = df.drop_duplicates()

# Replace 'Everyone' in 'Price' with '0'
df['Price'] = df['Price'].replace('Everyone', '0')

# Convert 'Installs' and 'Price' columns to numeric
df['Installs'] = pd.to_numeric(df['Installs'].replace('[\+,]', '', regex=True), errors='coerce')
df['Price'] = pd.to_numeric(df['Price'].replace('[\$,]', '', regex=True), errors='coerce')

# Analyze popularity, satisfaction, and pricing
popularity = df.groupby('Category')['Installs'].sum().sort_values(ascending=False)
satisfaction = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
pricing = df.groupby('Category')['Price'].mean().sort_values(ascending=False)

# Visualize insights
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
popularity.head(10).plot(kind='bar', color='grey')
plt.title('Popularity by Category')

plt.subplot(1, 3, 2)
satisfaction.head(10).plot(kind='bar', color='yellow')
plt.title('Satisfaction by Category')

plt.subplot(1, 3, 3)
pricing.head(10).plot(kind='bar', color='salmon')
plt.title('Average Price by Category')

plt.tight_layout()
plt.show()

# Recommendations
recommended_categories = popularity.head(3).index.tolist()
print(f"Recommended categories for a new app launch: {recommended_categories}")
```

### A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to launch a new app. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

### Using the Google Playstore statistics, I would examine important characteristics like upload dates, popular categories, user reviews, and download counts to provide recommendations on when to release new apps. Visualizations such as heatmaps, line charts, bar graphs, and scatter plots may be used to find patterns in user engagement and app activity. These trends can then be used to provide suggestions for the best time and emphasis for the app in order to optimize its performance in the market.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Playstore dataset from the provided link
url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
playstore_data = pd.read_csv(url)

# Convert 'Last Updated' column to datetime, handling errors
playstore_data['Last Updated'] = pd.to_datetime(playstore_data['Last Updated'], errors='coerce')

# Remove rows with NaT (Not a Time) values
playstore_data = playstore_data.dropna(subset=['Last Updated'])

# Line chart for app uploads over time
plt.figure(figsize=(10, 6))
playstore_data['Last Updated'].value_counts().sort_index().plot()
plt.title('App Uploads Over Time')
plt.xlabel('Date')
plt.ylabel('Number of App Uploads')
plt.show()

# Bar graph for category popularity
plt.figure(figsize=(10, 6))
playstore_data['Category'].value_counts().plot(kind='bar')
plt.title('Category Popularity')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=45)
plt.show()

# Scatter plot for user ratings and reviews
plt.figure(figsize=(10, 6))
plt.scatter(playstore_data['Rating'], playstore_data['Reviews'], alpha=0.5)
plt.title('User Ratings vs Reviews')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.show()


```
